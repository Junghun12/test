## 모델 선택 및 학습

### Subtask:

**Reasoning**:
Import the Linear Regression model from scikit-learn, define the independent and dependent variables, and train the model.



```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = df_height['fk'].values.reshape(-1, 1)
y = df_height['sk'].values

model = LinearRegression()
model.fit(X, y)
```

## 모델 예측

### Subtask:
학습된 모델을 사용하여 새로운 아빠 키에 대한 아들 키를 예측하는 함수 또는 코드를 제공합니다.


**Reasoning**:
Define a function to predict son's height using the trained model and provide an example of its usage.



```python
def predict_son_height(father_height):
  """
  Predicts the son's height based on the father's height using the trained linear regression model.

  Args:
    father_height: The height of the father (float or int).

  Returns:
    The predicted height of the son (float).
  """
  # Reshape the input to be a 2D array as required by the model
  father_height_reshaped = np.array(father_height).reshape(-1, 1)
  predicted_height = model.predict(father_height_reshaped)
  return predicted_height[0]

# Example usage: Predict the son's height for a father who is 175 cm tall
sample_father_height = 175
predicted_son_height = predict_son_height(sample_father_height)
print(f"Predicted son's height for a father of {sample_father_height} cm: {predicted_son_height:.2f} cm")
```
